## STEP 3

```python
from sklearn.model_selection import train_test_split
Xfill = xPoly
# Xfill = xLog
X_train, X_test, y_train, y_test = train_test_split(Xfill, y, test_size=1/3, random_state=13)
# X_train, X_test, y_train, y_test = train_test_split(Xfill, np.log10(y), test_size=1/3, random_state=13)
```

```python
from sklearn.linear_model import LinearRegression, ElasticNet, Lasso, Ridge
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
from sklearn.metrics import r2_score
```

------

## STEP 3

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
TestpredRegressor = regressor.predict(X_test)
```

```python
TrainpredRegressor = regressor.predict(X_train)
```

```python
print(f'Coefficients \t: {regressor.coef_}')
print('Intercept \t:', regressor.intercept_)
print('')
print("RMSE Test\t: %.2f" % mean_squared_error(y_test, TestpredRegressor, squared=False))
print("RMSE Train\t: %.2f" % mean_squared_error(y_train, TrainpredRegressor, squared=False))
print('')
print("MAE Test\t: %.2f" % mean_absolute_error(y_test, TestpredRegressor))
print("MAE Train\t: %.2f" % mean_absolute_error(y_train, TrainpredRegressor))
print('')
print('R2 Test score\t: %.2f' % r2_score(y_test, TestpredRegressor))
print('R2 Train score\t: %.2f' % r2_score(y_train, TrainpredRegressor))

print('')
print("train Accuracy : ",regressor.score(X_train,y_train))
print("test Accuracy : ",regressor.score(X_test,y_test))
```

```python
#RIGDE MODEL
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=3)
ridge_model.fit(X_train,y_train)
```
